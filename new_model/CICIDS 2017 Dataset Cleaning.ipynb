```python
# Load and combine data

import os
import pandas as pd

df = pd.DataFrame(columns=[' Destination Port', ' Flow Duration', ' Total Fwd Packets', ' Total Backward Packets', 'Total Length of Fwd Packets', ' Total Length of Bwd Packets', ' Fwd Packet Length Max', ' Fwd Packet Length Min', ' Fwd Packet Length Mean', ' Fwd Packet Length Std', 'Bwd Packet Length Max', ' Bwd Packet Length Min', ' Bwd Packet Length Mean', ' Bwd Packet Length Std', 'Flow Bytes/s', ' Flow Packets/s', ' Flow IAT Mean', ' Flow IAT Std', ' Flow IAT Max', ' Flow IAT Min', 'Fwd IAT Total', ' Fwd IAT Mean', ' Fwd IAT Std', ' Fwd IAT Max', ' Fwd IAT Min', 'Bwd IAT Total', ' Bwd IAT Mean', ' Bwd IAT Std', ' Bwd IAT Max', ' Bwd IAT Min', 'Fwd PSH Flags', ' Bwd PSH Flags', ' Fwd URG Flags', ' Bwd URG Flags', ' Fwd Header Length', ' Bwd Header Length', 'Fwd Packets/s', ' Bwd Packets/s', ' Min Packet Length', ' Max Packet Length', ' Packet Length Mean', ' Packet Length Std', ' Packet Length Variance', 'FIN Flag Count', ' SYN Flag Count', ' RST Flag Count', ' PSH Flag Count', ' ACK Flag Count', ' URG Flag Count', ' CWE Flag Count', ' ECE Flag Count', ' Down/Up Ratio', ' Average Packet Size', ' Avg Fwd Segment Size', ' Avg Bwd Segment Size', ' Fwd Header Length.1', 'Fwd Avg Bytes/Bulk', ' Fwd Avg Packets/Bulk', ' Fwd Avg Bulk Rate', ' Bwd Avg Bytes/Bulk', ' Bwd Avg Packets/Bulk', 'Bwd Avg Bulk Rate', 'Subflow Fwd Packets', ' Subflow Fwd Bytes', ' Subflow Bwd Packets', ' Subflow Bwd Bytes', 'Init_Win_bytes_forward', ' Init_Win_bytes_backward', ' act_data_pkt_fwd', ' min_seg_size_forward', 'Active Mean', ' Active Std', ' Active Max', ' Active Min', 'Idle Mean', ' Idle Std', ' Idle Max', ' Idle Min', ' Label'])

for dirname, _, filenames in os.walk('/Users/davidhan/Documents/HKU/Year_4/Final_Year_Project/research/cicids17/MachineLearningCVE'):
    for filename in filenames:
        path = os.path.join(dirname, filename)
        print("Loading data from", path)
        df_temp = pd.read_csv(path)
        df = df.append(df_temp, ignore_index = True)
        
df
```

```python
# Remove whitespace from column names 

df = df.rename(columns=lambda x: x.strip())

# Change column types to int64 (except label and those with float64)

cols = df.select_dtypes(include=['object']).columns.drop('Label')
df[cols] = df[cols].apply(pd.to_numeric, errors='coerce')
df.info()
```

```python
# Check original breakdown of classes

df['Label'].value_counts()
```

```python
# Check for rows with inf and NaN since they are useless and drop them

import numpy as np

df = df.replace([np.inf, -np.inf], np.nan)
df.isnull().sum()  

df = df.dropna()
```

```python
# Check value counts of certain attributes

print(df['Destination Port'].value_counts().head(10), '\n')

print(df['Flow Bytes/s'].value_counts().head(10), '\n')

print(df['Total Length of Bwd Packets'].value_counts().head(10), '\n')

print(df['Flow Packets/s'].value_counts().head(10), '\n')

print(df['Flow Duration'].value_counts().head(10), '\n')

print(df['FIN Flag Count'].value_counts().head(10), '\n')

print(df['Down/Up Ratio'].value_counts().head(10), '\n')

print(df['Bwd Avg Bytes/Bulk'].value_counts().head(10), '\n')
```

```python
# See breakdown for data types to see what needs to be corrected/normalized

pd.set_option('display.max_columns', 100)
df.describe()
```

```python
# Replace -1 in Init_Win_bytes_forward and Init_Win_bytes_backward with 0
# Explanation can be found in CIC flowmeter github issues (https://github.com/ahlashkari/CICFlowMeter/issues/21)

df['Init_Win_bytes_forward'] = df['Init_Win_bytes_forward'].replace(-1, 0)
df['Init_Win_bytes_backward'] = df['Init_Win_bytes_backward'].replace(-1, 0)
```

```python
# Check for rows with negative values since we shouldn't have any
# (i.e. flow duration, flow bytes/s, flow packets/s, fwd header length, etc.)

cols = df.columns.drop('Label')
df[cols].lt(0).sum().to_frame().T
```

```python
# Remove all rows with negative values in them where it doesn't make sense 

df.loc[df['Flow Duration'] < 0, 'Flow Duration'] = np.nan
df.loc[df['Flow Bytes/s'] < 0, 'Flow Bytes/s'] = np.nan
df.loc[df['Flow Packets/s'] < 0, 'Flow Packets/s'] = np.nan
df.loc[df['Flow IAT Mean'] < 0, 'Flow IAT Mean'] = np.nan
df.loc[df['Flow IAT Max'] < 0, 'Flow IAT Max'] = np.nan
df.loc[df['Flow IAT Min'] < 0, 'Flow IAT Min'] = np.nan
df.loc[df['Fwd IAT Min'] < 0, 'Fwd IAT Min'] = np.nan
df.loc[df['Fwd Header Length'] < 0, 'Fwd Header Length'] = np.nan
df.loc[df['Bwd Header Length'] < 0, 'Bwd Header Length'] = np.nan
df.loc[df['min_seg_size_forward'] < 0, 'min_seg_size_forward'] = np.nan

df = df.dropna()

cols = df.columns.drop('Label')
df[cols].lt(0).sum().to_frame().T
```

```python
# Divide port numbers into 3 categories:
# well known (0 to 1023), registered (1024 to 49151), or dynamic ports (49152 to 65535)
# To do this, we don't need to include dynamic ports as a column since it can be deduced from its absence from the former two
df['is_well_known_port'] = df.apply(lambda row: 1 if row['Destination Port'] < 1024 else 0, axis=1)
df['is_registered_port'] = df.apply(lambda row: 1 if row['Destination Port'] >= 1024 and row['Destination Port'] < 49152 else 0, axis=1)
```

```python
# Remove columns that aren't used

df = df.drop(columns = ['Destination Port', 'Fwd Header Length.1'], axis=1)
```
